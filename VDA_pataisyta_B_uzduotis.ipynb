# B. Kiek elektros energijos (kWh) vidutiniškai (mediana) per parą suvartoja vienas butas ir vienas namas kiekvieną metų mėnesį. Sukurkite lentelę ir liustraciją.

```python
# Iš bendros lenteles išsitraukiau visus duomenis susijusius su namu.

namas = duomenys.loc[(duomenys['OBT_PAVADINIMAS'] == 'Namas')]
namas
```

```python
# Susiradau kiek išviso yra namų, apie kuriuos yra gauta informacija.

namu_sk = len(namas['OBJ_NUMERIS'].unique())
namu_sk
```

```python
# Kodo pagalba radau informaciją apie tam tikrų metų tam tikrą mėnesį - kurią dieną mėnuo prasidėjo ir kiek tas mėnuo turi dienų.

from calendar import monthrange
monthrange(2020, 1)
```

```python
# Kodo pagalba pasiverčiau el. energios suvartojimo laiką į formatą, kurį atpažįsta programa.

namas["datetime"] = pd.to_datetime(namas['PL_T'], format='%Y-%m-%d %H:%M:%S')
namas['date'] = namas['datetime'].dt.date
namas['month'] = namas['datetime'].dt.month
```

```python
namu_menesiu_duomenys = []
for menuo in range(1, 13):
  # Atskira eilute kiekvienam namui. [namo_nr, sum_suvartojimas_ta_menesi]
  namo_suvartojimas = namas.loc[(namas.datetime.dt.month == menuo)].groupby(["OBJ_NUMERIS"])["P+"].sum()
  # Atskira eilute kiekvienam namui. [namo_nr, duomenu_dienu_skaicius]
  namo_suvartojimo_dienu_sk = namas.loc[(namas.datetime.dt.month == menuo)].groupby(["OBJ_NUMERIS"])["date"].nunique()
  # Atskira eilute kiekvienam namui. Vidutinis suvartojimas per diena is gautu duomenu
  namo_vidutinis_suvartojimas_per_diena = namo_suvartojimas / namo_suvartojimo_dienu_sk
  # Imamas vidurkis duomenu
  visu_namu_vidutinis_suvartojimas_per_men = namo_vidutinis_suvartojimas_per_diena.median()
  print(visu_namu_vidutinis_suvartojimas_per_men)
  namu_menesiu_duomenys.append(visu_namu_vidutinis_suvartojimas_per_men)
```

```python
# Išsitraukiau visą informaciją, susijusią su butais.

butas = duomenys.loc[(duomenys['OBT_PAVADINIMAS'] == 'Butas')]
butas
```

```python
# Gavau skaičių, apie kiek butų yra informacija.

butu_sk = len(butas['OBJ_NUMERIS'].unique())
butu_sk
```

```python
# Kodo pagalba pasiverčiau el. energijos suvartojimo laiką į programai atpažįstamą.

butas["datetime"] = pd.to_datetime(butas['PL_T'], format='%Y-%m-%d %H:%M:%S')
butas['date'] = butas['datetime'].dt.date
```

```python
butu_menesiu_duomenys = []
for menuo in range(1, 13):
  # Atskira eilute kiekvienam butui.
  buto_suvartojimas = butas.loc[(butas.datetime.dt.month == menuo)].groupby(["OBJ_NUMERIS"])["P+"].sum()
  # Atskira eilute kiekvienam butui. [buto_nr, duomenu_dienu_skaicius]
  buto_suvartojimo_dienu_sk = butas.loc[(butas.datetime.dt.month == menuo)].groupby(["OBJ_NUMERIS"])["date"].nunique()
  # Atskira eilute kiekvienam butui. Vidutinis suvartojimas per diena is gautu duomenu
  buto_vidutinis_suvartojimas_per_diena = buto_suvartojimas / buto_suvartojimo_dienu_sk
  # Imamas vidurkis duomenu
  visu_butu_vidutinis_suvartojimas_per_men = buto_vidutinis_suvartojimas_per_diena.median()
  print(visu_butu_vidutinis_suvartojimas_per_men)
  butu_menesiu_duomenys.append(visu_butu_vidutinis_suvartojimas_per_men)
```

```python
# ATSAKYMUI, kaip uždavinyje ir nurodo, sukūriau lentelę.
# Du stulpeliai skirti pamatyti kiek el. energijos suvartojo namai, ir kiek butai.
# Eilutėse išrašyti metų mėnesiai.

stulpeliai = ('NAMUOSE', 'BUTUOSE')
menesiai = ('Sausis','Vasaris','Kovas','Balandis','Gegužė','Birželis','Liepa','Rugpjūtis', 'Rugsėjis', 'Spalis', 'Lapkritis', 'Gruodis')

duomenys = {
    "MĖNESIAI": menesiai,
    "NAMUOSE": namu_menesiu_duomenys,
    "BUTUOSE": butu_menesiu_duomenys
}

lentele = pd.DataFrame(duomenys)
lentele = lentele.style.set_caption("Vidutinis el. energijos (mediana) suvartojimas per parą kWh")
lentele
```

```python
# ATSAKYME, kaip ir nurodyta uždavinyje, sukūriau grafiką vizualizacijai.
# Skirtingos linijos rodo būtų ir namų sunaudojamos el. energijos kitimą.

plt.plot(menesiai, butu_menesiu_duomenys, label = 'Butas', marker = '.', markersize = 10, linewidth = 0.3)
plt.plot(menesiai, namu_menesiu_duomenys, label='Namas', marker = '.', markersize = 10, linewidth = 0.3)

plt.legend()

plt.xlabel('mėnuo')
plt.ylabel('kWh per parą')

plt.xticks(rotation=30, ha='right')
plt.title('Vidutinis el. energijos suvartojimas')

plt.show()
```
